```python
import json
import re

def process_string(input_string, ap=True):
    result = {}
    
    # Split the string into lines
    lines = input_string.strip().split("\n")
    
    for line in lines:
        # Regular expression to capture the relevant parts
        if ap:
            match = re.match(r'([0-9.]+)% Accuracy\s+([a-zA-Z0-9._]+)\s+\((\d+)\)\s+\[.*\]', line)
        else:
            match = re.match(r'([0-9.]+)% F1\s+([a-zA-Z0-9._]+)\s+\((\d+)\)', line)
        
        if match:
            # Extract accuracy, question_id, and the number inside parentheses
            accuracy = float(match.group(1)) / 100  # Normalize accuracy to a value between 0.0 and 1.0
            question_id = match.group(2)
            frequency = round(int(match.group(3)) / 60, 4)  # Divide the number inside parentheses by 60
            
            # Populate the result dictionary
            result[question_id] = {
                "question_confidence": accuracy,
                "question_frequency": frequency
            }
    
    return result

# Call the function and print the result
answer_prediction = {
    "bg": process_string(bg),
    "el": process_string(el),
    "en": process_string(en),
    "pl": process_string(pl),
    "ro": process_string(ro),
}
evidence_extraction = {
    "bg": process_string(bg_ee, False),
    "el": process_string(el_ee, False),
    "en": process_string(en_ee, False),
    "pl": process_string(pl_ee, False),
    "ro": process_string(ro_ee, False),
}
all_results = {
    "answer_prediction": answer_prediction,
    "evidence_extraction": evidence_extraction
}
print(json.dumps(all_results, indent=4))
```
